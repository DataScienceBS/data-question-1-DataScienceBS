#### Question 22a:  Are there differences across years?  
##### Answer 22a: Yes, the percentage of Internet users is increasing from 2002 to 2007 to 2012. 

#### Question 22b: What do the plots tell you about any relationship between these two variables?
##### Answer 22b: Having all three years on the same scatter plot makes it easy to identify the shift in outliers from one year to the next, but does not associate colors with year, so the assumption is made that it has increased. Having the three separate scatter plots shows a growth in bins from one year to the next, expanding from an 80% max to 100% max in the display, making it easy to identify growth in Percentage of Internet Users. However, the majority of the data lies alone the x-axis and the outliers don't seem to correlate to the y-axis. I don't feel these two variables have a relationship that is communicated through a scatter-plot.
